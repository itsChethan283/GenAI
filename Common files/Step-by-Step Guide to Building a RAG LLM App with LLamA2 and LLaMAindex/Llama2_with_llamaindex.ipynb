```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
from IPython.display import Markdown, display
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import Simple
```

```python
docs = SimpleDirectoryReader("/content/Data").load_data()
# docs
```

```python
system_prompt = """
You are a Q/A assistant. Your goal is to answer as
precisely as possible based on the instructions and context provided.
"""
query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(context_window=4096, max_new_tokens=256,
                     query_wrapper_prompt=query_wrapper_prompt,
                     system_prompt=system_prompt,
                     generate_kwargs={"temperature": 0.0, "do_sample": False},
                     tokenizer_name="daryl149/llama-2-7b-chat-hf",
                     model_name="daryl149/llama-2-7b-chat-hf",
                     device_map="auto",
                     model_kwargs={"torch_dtype": torch.torch.float16, "load_in_8bit": True})
```

```python
from langchain_community.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding


embed_model = LangchainEmbedding(HuggingFaceEmbeddings())
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size = 1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
service_context
```

```python
index = VectorStoreIndex.from_documents(docs, service_context=service_context)
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What is attention is all you need?")
```
